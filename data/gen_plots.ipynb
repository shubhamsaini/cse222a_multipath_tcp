## Test MPTCP Connection

```r
filename="test_connection"
data = read.csv(paste0(filename,".csv"))
data$Graph.1 = data$Graph.1/1024
data$Graph.2 = data$Graph.2/1024

ggplot(data, aes(Interval.start)) + 
  geom_line(aes(y = Graph.1, colour = "eth1")) + 
  geom_line(aes(y = Graph.2, colour = "eth2")) +
  ylim(1200,1800) +
  xlab("Seconds") +
  ylab("KB/sec") + coord_fixed(ratio=0.05) +
theme_bw()

ggsave(paste0(filename,".pdf"))
```

## Bandwidth Aggregation

```r
filename="band_agg"
data = read.csv(paste0(filename,".csv"))
data$Graph.1 = data$Graph.1/1024
data$Graph.2 = data$Graph.2/1024
data$total = data$Graph.1 + data$Graph.2

ggplot(data, aes(Interval.start)) + 
  geom_line(aes(y = Graph.1, colour = "eth1")) + 
  geom_line(aes(y = Graph.2, colour = "eth2")) +
  geom_line(aes(y = total, colour = "total")) +
  xlab("Seconds") +
  ylab("KB/sec") + coord_fixed(ratio=0.005) +
theme_bw()

ggsave(paste0(filename,".pdf"))
```

## HTSIM

```r
library('lattice')
trellis.par.set(col.whitebg())
source('readdata.R')
res <- parseTrace("fully.dat")
res = as.data.frame(res)
pktsize <- attr(res,'pktsize')
servperflow <- attr(res,'bottleneckrate')/attr(res,'numflows')

rescale <- res$type %in% c('queueapprox','tcp','tcprecord')
res$val1[rescale] <- res$val1[rescale]/pktsize
res$val2[rescale] <- res$val2[rescale]/pktsize
res$val3[rescale] <- res$val3[rescale]/pktsize

idtype <- res$id %in% c('Tcp1','Subflow1','Subflow2')
res = res[idtype,]

res$id = droplevels(res$id)

resbak = res

my.plot <- xyplot(val1~time, data=resbak, group=id, subset= (type=='tcp' | type=='tcprecord' |
         type=='queueapprox') & (id!='ratelim%1') &
         time>0 & time<50, type='l',
      auto.key=list(space="top", columns=4, 
                       title="District", cex.title=1), xlab="Seconds", ylab="cwnd")

#trellis.device(device="png", filename="xyplot.png")
#print(my.plot)
#dev.off()

print(my.plot)
```

```r
filename="fair_2/tcp_balia"
data = read.csv(paste0(filename,".csv"))
names(data) = c("time", "tcp")
data$time <- as.POSIXct(strptime(data$time, format="%H:%M:%S"))
data$tcp = data$tcp/1024

filename="fair_2/mptcp_balia"
data2 = read.csv(paste0(filename,".csv"))
names(data2) = c("time", "mptcp1", "mptcp2")
data2$time <- as.POSIXct(strptime(data2$time, format="%H:%M:%S"))
data2$mptcp1 = data2$mptcp1/1024
data2$mptcp2 = data2$mptcp2/1024

mdata = merge(data, data2, all = T)

ggplot(mdata, aes(x = time)) + 
  geom_line(aes(y = tcp, colour = "tcp")) + 
  geom_line(aes(y = mptcp2, colour = "mptcp")) + 
  xlab("Seconds") +
  ylab("KB/sec") + coord_fixed(ratio=0.05) +
theme_bw()

ggsave(paste0(filename,".pdf"))
```
